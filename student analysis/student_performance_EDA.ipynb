# univariate analysis

```python
sns.histplot(df["study_method"])
sns.show()
```

```python
sns.histplot(df["final_grade"])
sns.show()
```

# bivariate analysis

```python
sns.barplot(x="study_hours",y="final_grade",data=df)
plt.show()
```

```python
sns.lineplot(y="attendance_percentage",x="final_grade",data=df)
plt.show()
```

```python
sns.heatmap(df[["math_score",
"science_score",
"english_score","overall_score" ,          
"final_grade"]].corr(),annot=True)

plt.title("score correletion")
plt.show()
```

```python
sns.pairplot(df[["math_score",
"science_score",
"english_score"]])
plt.show()
```

Math , Science , English scores shows a positive relationship , meaning students who score high in one subject tend to score high in other subjects
